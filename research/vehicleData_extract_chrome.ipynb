```python
def setup_API_env(user_agents):


    # Define the headers
    
    user_agent = random.choice(user_agents)
    headers = {
    "Accept": "*/*",
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Accept-Language": "en-US,en;q=0.9",
    "Authorization": "Basic Mjk4OTI0OnBiczkyNA==",
    "Connection": "keep-alive",
    "Content-Type": "application/json",
    "Origin": "https://demos.autodatasolutions.com",
    "Referer": "https://demos.autodatasolutions.com/ADSDemo/",
    "Sec-Fetch-Dest": "empty",
    "Sec-Fetch-Mode": "cors",
    "Sec-Fetch-Site": "same-origin",
    "User-Agent": user_agent
    }


    # Define the payload (if any)
    payload = {"languageLocale":"CA_en"}

    api_gateway_search_by_model = f"https://demos.autodatasolutions.com/ADSDemo/api/full/ymmt/year/2020/make/Ford/model/F-150/"

    api_gateway_search_by_vin = f"https://demos.autodatasolutions.com/ADSDemo/api/full/vindecode/"
    return headers, payload, api_gateway_search_by_vin, api_gateway_search_by_model
```

```python
# api_gateway = "https://demos.autodatasolutions.com/ADSDemo/api/full/ymmt/year/2020/make/Ford/model/F-150/"


headers, payload, api_gateway_search_by_vin,api_gateway_search_by_model  = setup_API_env(user_agents)
```

```python

search_results_by_model = requests.post(api_gateway_search_by_model, headers=headers, json=payload)
search_results_by_vin = requests.post(api_gateway_search_by_vin +"1FTEW1EP6LFC83313", headers=headers, json=payload)
```
